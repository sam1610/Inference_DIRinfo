```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder
```

```python
df=pd.read_csv("data/GD_infer.csv")
df.head()
```

```python
#df=df[df.SUBM=="Math"] 
df[["GPA", "NAT_MARK"]].T
```

```python
df.drop([ 
        'TOTAL1','ACHIEVEMENTS2','MIDTERM2', 'FINAL2', 'TOTAL2', 'D_MARK', 'Unnamed: 0'], axis=1, inplace=True)
```

```python
df.set_index("CPR_NO", inplace=True)
```

```python
df.head()
```

```python
df.keys()
```

```python
df=df[['NAT_class', 'GPA_class','SCH_NM', 'PROV_NM', 'SUBM', 'Teacher_Gender', 'ACHIEVEMENTS1',
       'MIDTERM1', 'FINAL1', 'NAT_MARK', 'GPA' ]]
```

```python

df.head(10)
```

```python
df.NAT_class.unique()
```

```python
df.GPA_class.unique()
```

```python
df.GPA_class.value_counts()
```

```python
df.GPA_class=df.GPA_class.map(
{
    '<60':0,'60-70':1,  '70-80':2 , '80-90':3, '>90':4} )
```

```python
df.NAT_class=df.NAT_class.map(
{
    '<60':0,'60-70':1,  '70-80':2 , '80-90':3, '>90':4}  
)
```

```python
le = LabelEncoder()

# for i in ['SCH_NM','Teacher_Gender','GPA_class','NAT_class' ]:

#     df[i] = label_encoder.fit_transform(df[i])
# le=LabelEncoder()
df.SCH_NM=le.fit_transform(df.SCH_NM)
```

```python
df.Teacher_Gender=((df.Teacher_Gender=="M")+0)
```

```python
df.SUBM=((df.SUBM=="Math")+0)
```

```python
df.head(10)
```

```python
df= pd.get_dummies(df, columns=["SCH_NM", "PROV_NM"])
```

```python
df.to_csv("data/aws_data_classification.csv", header=False, index=False)
```
